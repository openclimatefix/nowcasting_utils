```python
import pandas as pd
import plotly.graph_objects as go

# make histrogram to compare two different models runs
# will be comparing NMAE at a GSP level

file_1 = 'OptiFlow PRED-714_normalized.csv'
file_2 = 'ESO forecast_normalized.csv'

ocf_df = pd.read_csv(file_1)
eso_df = pd.read_csv(file_2)

print(ocf_df.columns)
```

```python
# plot 
colours = ["rgb(77,137,99)", "rgb(225,179,120)", "rgb(224,204,151)", "rgb(105,165,131)"]


xbins=dict(
        start=-1,
        end=30,
        size=1
    )

trace_ocf = go.Histogram(
    x=ocf_df["MAE"],
    marker_color=colours[0],
    name='OCF Optiflow',
    xbins=xbins
)

trace_eso = go.Histogram(
    x=eso_df["MAE"],
    marker_color=colours[1],
    name='NG ESO',
    xbins=xbins
)

fig = go.Figure()
fig.add_trace(trace_ocf)
fig.add_trace(trace_eso)

fig["layout"]["xaxis"]["title"] = "GSP NMAE [%]"
fig["layout"]["yaxis"]["title"] = "Count"
fig["layout"]["title"] = "Histogram to compare GSP NMAE error of NG ESO and OCF models"
fig.update_xaxes(range = [0,30])

fig.show()

    
```
